```python
# FIXED: STOCK PRICE PREDICTION - MULTI-FILE VERSION WITH ERROR HANDLING

import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Path to the folder containing stock CSVs
folder_path = "/content/sample_data/stock dataset"

# List of CSV files
file_list = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Loop through files
for file in file_list:
    print(f"\nüîé Processing file: {file}")

    # Load the file
    df = pd.read_csv(os.path.join(folder_path, file))

    # Check for 'Close' column
    if 'Close' not in df.columns:
        print("‚ùå Skipping: 'Close' column not found.")
        continue

    # Create 'Next_Close' target and drop NaNs
    df['Next_Close'] = df['Close'].shift(-1)
    df.dropna(inplace=True)

    # Skip if not enough data
    if len(df) < 10:
        print(f"‚ö†Ô∏è Skipping: Not enough data points ({len(df)} rows).")
        continue

    # Features and target
    X = df[['Close']]
    y = df['Next_Close']

    # Train/test split
    try:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
    except ValueError as e:
        print(f"‚ö†Ô∏è Skipping due to error in train_test_split: {e}")
        continue

    # Train the model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Predict
    y_pred = model.predict(X_test)

    # Evaluate
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(f"üìâ Mean Squared Error: {mse:.2f}")
    print(f"üìà R¬≤ Score: {r2:.2f}")

    # Plot
    plt.figure(figsize=(10, 4))
    plt.plot(y_test.values, label='Actual', color='blue')
    plt.plot(y_pred, label='Predicted', color='red', linestyle='--')
    plt.title(f"{file} - Actual vs Predicted Closing Prices")
    plt.xlabel("Days")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
